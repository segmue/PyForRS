### Notebooks

The body of a notebook is composed of cells. Each cell contains either markdown (a common markup language), code input, code output, or raw text. Cells can be included in any order and edited at-will, allowing for a large amount of flexibility for constructing a narrative.

- **Markdown cells** - These are used to build a nicely formatted narrative around the code in the document. The majority of this lesson is composed of markdown cells.

- **Code cells** - These are used to define the computational code in the document. They come in two forms: the *input cell* where the user types the code to be executed, and the *output cell* which is the representation of the executed code. Depending on the code, this representation may be a simple scalar value, or something more complex like a plot or an interactive widget.

- **Raw cells** - These are used when text needs to be included in raw form, without execution or transformation.

### Integer and Float
An *integer* is as a number that can be written without a fractional component. Examples are 1, 42, -9999, and 12345. You can assign integers to a name and use mathematical operations on them:

`````python
int1 = 1
int2 = 2
int3 = int1 + int2
int3
`````

Note that the *type* of the sum of two integers is still an integer:

`````python
type(int3)
`````

A *floating point number* (or *float* in short) is a number that can include a fractional component. Examples are 1.0, -4.4, 3.14, 1.2 $\times$ 10$^9$ and the mathematical constant $\pi$. In contrast to integers, floating point numbers have a limited precision, because not all decimal places can be stored in the computer's memory.

Here a some examples with floats

`````python
float1 = 1.0
float2 = 2.0
float3 = float1 + float2
float3
`````

`````python
type(float3)
`````

`````python
large_number = 1.2e9
print(large_number)
`````

Note that when you divide to integers, you might get a float as results (e.g. `1 / 2 = 0.5`), because the number 0.5 cannot be stored as integer object:

`````python
c = int1 / int2
print(c, type(c))
`````

## Control Flow

To control the flow of your program, you need the possibility to (a) execute some parts of the code only if some condition is true and (b) repeat some part of the code.

For the first part, you can use the `if` and `else` keywords. Consider the following example: 

`````python
condition = False

if condition:
    print("Condition is true")
    print("Still on the IF block")
else:
    print("Condition is false")
print("Now we are out!")
`````

The first thing we need to know is that Python uses identation to create code blocks (sections of code that belong together). Everything behind the colon `:` that is indented is only executed if the condition is true. See what happens if you change the `condition` to `False`.

In Python, indentation should always use four spaces per indentation level. In Jupyter Notebooks, you can still press the `Tab` key, which is automatically replaced with four spaces.

To repeat some part of the code, consider the example below. Here we see that the code is executed _while_ a certain condition evaluates to true. We call this a _while loop_. The _while_ keyword is followed by a condition and for as long as the condition is true, it will run. Try for example to change the condition from "> 1" to "< 11" and see what happens. 

<par>
    
Luckily we have the variable _count_ that _breaks_ our execution when it reaches the value 20, otherwise the code would execute forever. This brings us to the _if_ statements, _if_ evalutes a condition, and if it is true, it will display or do what ever we write in its body. The _if_ statement can be coupled with the _else_ statement, which is executed in cases where _if_ evaluated to be false. Here we check weather a given _value_ can be divided by 2 without remainder (the so-called modulo operator %). Finally the _break_ statement will stop the whole execution of the code block (the _while_ loop in this case).

`````python
value = 25
count = 0
while value >= 0:
    if value % 2 == 0:
        print("value", value, "is even")
    else :
        print("value", value, "is not even")
    value -= 1
    count += 1
    if count == 20:
        break
`````

The most common loop is the _for_ loop, which iterates over an iterable object, for example, a list or a dictionary:

`````python
for number in ["1", "2", "3"]:
    print("Sentinel -", number)
`````

A _for_ loop is often combined with the `range(start, stop, step)` function, which iterates from `start` to `stop` with a step size of `step`. Note that `stop` is not printed: 

`````python
start = 0
stop = 10
step = 1

for i in range(start, stop, step):
    print(i)
`````

Besides the `break` statement, Python also has the `continue` statement which will jump straight to the next iteration. Look at the code and think about the output you expect. 

`````python
for i in range(10):
    print("i = ", i, " is...")
    if i % 2 == 0:
        print(" even")
        continue
    else:
        print(" uneven")
    print("goodbye")
`````

### String formatting
To convert strings to numbers, you can use the `int` or `float` function. To convert numbers to strings, you use the `str` function or the `str.format` method, which provides the additional option to format you string. Here, string contains place holders identified by braces (`{` or `}`) for substituting the numbers:

`````python
'{foo} and {bar}'.format(foo=42, bar=12)
`````

You can use a colon `:` followed by a format the number (see https://docs.python.org/3/library/string.html#formatstrings for details): 

`````python
print('{foo:.1f}'.format(foo=3.16))  # format to floating point number with one digit behind the dot.
print('{foo:+5d}'.format(foo=42))    # format to decimal and forcing plus or minus sign
print('{foo:.2e}'.format(foo=3.16))  # format to exponential notation with two digits
`````

### Line plots

Let's plot a simple graph of cosine(x) and sine(x), where the array x has 256 values between $-\pi$ and $\pi$.

`````python
x = np.linspace(-np.pi, np.pi, 256, endpoint=True) #np.pi --> get the value of pi from the numpy library
c = np.cos(x)
s = np.sin(x)

# create a figure and axes object
fig, ax = plt.subplots()

# plot cosine and sine
ax.plot(x, c, label = 'cosine')
ax.plot(x, s, label = 'sine')

# add labels
ax.set_xlabel('x')
ax.set_ylabel('y(x)')

# set x and y limits of axes
ax.set_xlim(-4, 4)
ax.set_ylim(-1,1)

# add grid
ax.grid(True)

# add a legend
ax.legend()
`````

All methods of the axes used here have several options that you can explore. The axes also has many more methods than used here. To learn more about Matplotlib, you can also have a look a the [Matplotlib Tutorial](https://matplotlib.org/tutorials/index.html) and the [Matplotlib Gallery](https://matplotlib.org/gallery/index.html) with many code examples.
